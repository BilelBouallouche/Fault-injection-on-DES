```python
def permutation(bits, perm, perm_taille):
    """
        permute la chaine de caractères bits suivant le tableau de permutation perm, de taille perm_taille
    """
    res = []
    for i in range(0, perm_taille):
        p = perm[i]
        res.append(bits[p-1])
    return ''.join(res)

def inverse_perm(perm, taille_finale):
    """
        Inverse une permutation perm
    """
    res = []
    for i in range(1, len(perm)+1):
        j = 0
        while (j < len(perm)) and (perm[j] != i):
            j+=1
        if len(res) < taille_finale:
            res.append(j+1)
    return res

def expansion(R_im1):
    """
        Renvoie Expansion(R_im1), une des opérations faites dans la fonction de tour du schéma de Feistel 
    """
    return permutation(R_im1, E, 48)


IPinverse = inverse_perm(IP, 64)
Pinverse = inverse_perm(P, 32)
```

```python
def xor(a, b):
    """
        retourne a XOR b
        où a et b sont des chaines de caracteres representant des nombres binaires
    """
    return bin(int(a, 2) ^ int(b, 2))[2:].zfill(len(a))
```

```python
import itertools
def generer_combinaisons_bits(n):
    """
    génère l'ensemble des chaines de caractères représentant un nombre binaire de n bits 
    """
    return [''.join(seq) for seq in itertools.product("01", repeat=n)]

K16s = generer_combinaisons_bits(6)
```

```python
def appliquer_sbox(exp_R, s):
    """
        Applique a exp_R la sbox numéro s
    """
    b0 = str(exp_R[0])
    b1 = str(exp_R[1])
    b2 = str(exp_R[2])
    b3 = str(exp_R[3])
    b4 = str(exp_R[4])
    b5 = str(exp_R[5])
    ligne = int(b0+b5, 2)
    colonne = int(b1+b2+b3+b4, 2)
    return bin((sboxs[s][ligne])[colonne])[2:].zfill(4)
```

```python
def inter(Ls):
    """
        Renvoie l'élément i qui se trouve dans toutes les listes stockés dans la liste de listes Ls
    """
    res = set(Ls[0])
    for e in Ls[1:]:
        res.intersection_update(e)
    return list(res)[0]
```

```python
def attaquer_sbox():
    """
        Fonction de recherche exhaustive sur les sboxs.
        Son fonctionnement est décrit dans le rapport
    """
    candidat_par_sbox = {i: [] for i in range(1, 9)}
    res = []
    faux_par_sbox = trouver_fautes()
    
    for sbox, faux in faux_par_sbox.items():
        for f in faux:
            ant_faux = permutation(f, IP, 64) 
            #L16_faux = liste_L16_faux[faux]
            L16_faux, R16_faux = ant_faux[0:32], ant_faux[32:]
            R15_faux = R16_faux
            R15_faux_exp = expansion(R15_faux)
            delta = xor(L16, L16_faux)
            y = permutation(delta, Pinverse, 32)
            sols = []
            for k16 in K16s:
                for b in range(0, 8): 
                    ant_juste = xor(k16, R15_exp[b*6:(b+1)*6])
                    ant_faux = xor(k16, R15_faux_exp[b*6:(b+1)*6])
                    sortie_sbox_juste = appliquer_sbox(ant_juste, sbox)
                    sortie_sbox_faux = appliquer_sbox(ant_faux, sbox)
                    comp = xor(sortie_sbox_juste, sortie_sbox_faux)
                    if (comp == y[b*4:(b+1)*4]) and (comp != '0000'):
                        sols.append(int(k16, 2))
            candidat_par_sbox[sbox].append(sols)
        solution = inter(candidat_par_sbox[sbox])
        res.append(bin(solution)[2:].zfill(6))
    return res

attaquer_sbox()
```

```python
def retrouver_K16(morceaux_k16):
    """
        On reconstitue K_{16} par simple concaténation des portions trouvés avec l'attaque sur les sboxs
    """
    k16 = ""
    for m in morceaux_k16:
        k16 += m
    return k16

k16 = retrouver_K16(attaquer_sbox())
print("Valeur de K_{16} :")
print("--- en binaire --| ", k16)
print("---- en hexa ----| ",hex(int(k16, 2)))
```

```python
import requests
from bs4 import BeautifulSoup

def DES_en_ligne(k_test):
    """
        Fonction qui verifie si la clé a tester K_test est la bonne
        en verifiant qu'elle permet bien de déchiffrer le chiffré juste en obtenant le clair
        On se sert ici de https://emvlab.org/descalc/ comme calculatrice DES.
        En effet, si l'on va sur le site et qu'on l'utilise, on se rend compte que les données que l'on rentre
        sont transmises au serveur directement via l'URL sous la forme:
        https://emvlab.org/descalc/key=[CLE]'&iv=0000000000000000&input=[CLAIR]'&mode=ecb&action=Encrypt&output='
        On peut alors se servir du package requests de Python pour faire au serveur une requête URL 
        avec les données de notre probleme 
        On inspecte ensuite le code HTML du site et on voit que la zone de texte est balisée par
        <textarea cols="48" id="output" name="output" rows="10"></textarea></p>
        On peut donc se servir du package BeautifulSoup pour lire la réponse dans la textearea "Output data"
        (avec comme id "output") du site
    """
    
    url = "https://emvlab.org/descalc/?"
    url += 'key=' + k_test + '&iv=0000000000000000&input=' + hex(int(clair,2))[2:] + '&mode=ecb&action=Encrypt&output='

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    if soup.find(id='output').getText() == hex(int(chiffre_juste,2))[2:].upper():
        return True
    return False
```

```python
def positions_manquantes():
    """
        Trouve les positions manquantes dans la clé
        Consiste en un simple parcourt de chaîne de caractères
    """
    res = []
    for i in range(0, len(k_tmp)):
        if k_tmp[i] == '*':
            res.append(i)
    return res
```

```python
def completer_k():
    """
        Attaque par recherche exhaustive pour completer la clé K en retrouvant les 8 bits manquants
    """
    positions_manq = positions_manquantes()
    possibilites = generer_combinaisons_bits(8)
    for combi in possibilites:
        #if combi != '00000000':
        k_test = []
        k_test[:0] = k_tmp
        for i, b in enumerate(combi):
            k_test[positions_manq[i]] = b
        k_test = ''.join(k_test)
        if(DES_en_ligne(hex(int(k_test,2))[2:].upper())):
            return k_test, hex(int(k_test,2))[2:]
```

```python
def parite(k):
    """
        Reecriture des octets composants la clé de telle sorte a ce que chacun ait un nombre impair de 1.
         Il suffit de compter le nombre de 1 et de 0 a chaque octet et de réécrire le dernier bit
    """
    k64 = ""
    for i in range(0, 64, 8):
        octet = []
        octet[:0] = k[i:i+8]
        if octet.count('1') % 2 == 0:
            octet[-1] = '1'
        else:
            octet[-1] = '0'
        k64 += ''.join(octet)
    return k64
```
